```python
import pandas as pd
import numpy as np
import matplotlib as plt
from tensorflow.keras.models import load_model
import pickle
```

```python
with open('label_encoder_gender.pkl', 'rb') as file:
    label_encoder = pickle.load(file)
with open('one_hot_encoder_geo.pkl', 'rb') as file:
    ohe_encoder = pickle.load(file)

with open('scaler.pkl', 'rb') as file:
    scaler = pickle.load(file)


model = load_model('model.h5')
```

```python
input_data = {
    'CreditScore':600,
    'Geography':'France',
    'Gender':'Male',
    'Age':40,
    'Tenure':3,
    'Balance':60000,
    'NumOfProducts':2,
    'HasCrCard':1,
    'IsActiveMember':1,
    'EstimatedSalary':50000
}
```

```python
input_data['Gender'] = label_encoder.fit_transform(['Gender'])[0]
```

```python
input_data
```

```python
geo_encoder = ohe_encoder.transform([[input_data['Geography']]]).toarray()
geo_encoded_df = pd.DataFrame(geo_encoder, columns=ohe_encoder.get_feature_names_out(['Geography']))
geo_encoded_df
```

```python
ip = pd.DataFrame([input_data])
```

```python
df = pd.concat([ip.drop('Geography', axis=1), geo_encoded_df], axis=1)
df
```

```python
input_scaled = scaler.transform(df)
input_scaled
```

```python
prediction = model.predict(input_scaled)
prediction[0][0]
```
